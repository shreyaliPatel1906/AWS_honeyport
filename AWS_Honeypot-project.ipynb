```python
if 'srcstr' in df.columns:
    plt.figure(figsize=(10, 5))
    df['srcstr'].value_counts().head(10).plot(kind='bar', color='skyblue')
    plt.title('Top 10 Source IPs')
    plt.xlabel('Source IP')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
if 'dpt' in df.columns:
    plt.figure(figsize=(10, 5))
    df['dpt'].value_counts().head(10).plot(kind='bar', color='salmon')
    plt.title('Top 10 Destination Ports')
    plt.xlabel('Port')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
if 'proto' in df.columns:
    plt.figure(figsize=(8, 5))
    df['proto'].value_counts().plot(kind='bar', color='lightgreen')
    plt.title('Protocol Distribution')
    plt.xlabel('Protocol')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# 4. Activity Over Time
if 'datetime' in df.columns:
    plt.figure(figsize=(12, 5))
    df['datetime'].value_counts().sort_index().plot()
    plt.title('Activity Over Time')
    plt.xlabel('Date')
    plt.ylabel('Number of Events')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
if 'dpt' in df.columns and 'proto' in df.columns:
    heatmap_data = df.groupby(['proto', 'dpt']).size().unstack(fill_value=0)
    plt.figure(figsize=(12, 6))
    sns.heatmap(heatmap_data, cmap='YlGnBu')
    plt.title('Heatmap: Protocol vs Destination Port')
    plt.xlabel('Destination Port')
    plt.ylabel('Protocol')
    plt.tight_layout()
    plt.show()
```
